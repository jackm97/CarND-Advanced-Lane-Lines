```python
reload(img_manip)
test_images = os.listdir('test_images/')
num = len(test_images)

plt.figure(figsize=(20,60))
for idx, fname in enumerate(test_images):
    image = cv2.imread('test_images/'+ fname)
    dst = img_manip.undistort(image)
    
    s1,_ = img_manip.hsv_thresh(image) # saturation channel of HSV colorspace
    s2,_ = img_manip.hls_thresh(image) # saturation channel of HLS colorspace
    
    # Threshold HSV saturation gradients
    gradx1,gradx1_thresh     = img_manip.sobelx_thresh(s1,(20,80))
    grady1,grady1_thresh     = img_manip.sobely_thresh(s1,(30,120))    
    gradxy1,gradxy1_thresh   = img_manip.sobelxy_thresh(gradx1,grady1,(40,100)) # magnitude
    graddir1,graddir1_thresh = img_manip.sobeldir_thresh(gradx1,grady1,(60,110)) # direction
    
    combined_thresh1 = np.zeros_like(s1)
    combined_thresh1[(gradx1_thresh & grady1_thresh).astype(bool) | (gradxy1_thresh & graddir1_thresh).astype(bool)] = 1
    
    # Threshold HLS saturation gradients
    gradx2,gradx2_thresh     = img_manip.sobelx_thresh(s2,(20,80))
    grady2,grady2_thresh     = img_manip.sobely_thresh(s2,(30,150))    
    gradxy2,gradxy2_thresh   = img_manip.sobelxy_thresh(gradx2,grady2,(50,150))  # magnitude
    graddir2,graddir2_thresh = img_manip.sobeldir_thresh(gradx2,grady2,(50,100)) # direction
    
    combined_thresh2 = np.zeros_like(s2)
    combined_thresh2[(gradx2_thresh & grady2_thresh).astype(bool) | (gradxy2_thresh & graddir2_thresh).astype(bool)] = 1 
    
    
    plt.subplot(num,2,1+2*idx)
    plt.imshow(combined_thresh1,cmap='gray')
    plt.title('HSV')
    plt.subplot(num,2,2+2*idx)
    plt.imshow(combined_thresh2,cmap='gray')
    plt.title('HLS')
    plt.show()
```

```python
def get_warped_binary(image):
    image = cv2.imread('test_images/'+ fname)
    dst = img_manip.undistort(image)
    
    # Color Threshold
    s1,color_thresh1 = img_manip.hsv_thresh(dst) # saturation channel of HSV colorspace
    s2,color_thresh2 = img_manip.hls_thresh(dst) # saturation channel of HLS colorspace
    
    color_thresh = np.zeros_like(color_thresh1)
    color_thresh[(color_thresh1.astype(bool) | color_thresh2.astype(bool))] = 1
    
    # Threshold HSV saturation gradients
    gradx1,gradx1_thresh     = img_manip.sobelx_thresh(s1,(20,80))
    grady1,grady1_thresh     = img_manip.sobely_thresh(s1,(30,120))    
    gradxy1,gradxy1_thresh   = img_manip.sobelxy_thresh(gradx1,grady1,(40,100)) # magnitude
    graddir1,graddir1_thresh = img_manip.sobeldir_thresh(gradx1,grady1,(60,110)) # direction
    
    combined_thresh1 = np.zeros_like(s1)
    combined_thresh1[(gradx1_thresh & grady1_thresh).astype(bool) | (gradxy1_thresh & graddir1_thresh).astype(bool)] = 1
    
    # Threshold HLS saturation gradients
    gradx2,gradx2_thresh     = img_manip.sobelx_thresh(s2,(20,80))
    grady2,grady2_thresh     = img_manip.sobely_thresh(s2,(30,150))    
    gradxy2,gradxy2_thresh   = img_manip.sobelxy_thresh(gradx2,grady2,(50,150))  # magnitude
    graddir2,graddir2_thresh = img_manip.sobeldir_thresh(gradx2,grady2,(50,100)) # direction
    
    combined_thresh2 = np.zeros_like(s2)
    combined_thresh2[(gradx2_thresh & grady2_thresh).astype(bool) | (gradxy2_thresh & graddir2_thresh).astype(bool)] = 1
    
    binary = np.zeros_like(combined_thresh1)
    binary[(color_thresh.astype(bool) | (combined_thresh1.astype(bool) | combined_thresh2.astype(bool)))] = 1
    
    warped_binary = img_manip.warp(binary)
    
    return warped_binary
```
